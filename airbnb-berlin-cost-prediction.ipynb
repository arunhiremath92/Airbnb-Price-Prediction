```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = None
df = pd.read_csv("./berlin-airbnb-data/listings_summary.csv")
```

```python
df.describe(include="all")
```

```python
#first lets get started and remove all the data that DEFINITELY not needed for analysis
#like the listing_url,scraped data, scraped id, name
#Also trying to remove any sections that has summary in it, since we are not going to include Text-Processing based results in prediction
##Check if reviews could be used dropping them for now,
replace_dollar_symbols = True
drop_columns = True
```

```python
df["amenities"]
```

```python
columns_to_be_deleted = ["listing_url", "host_total_listings_count", "scrape_id", "experiences_offered", "last_scraped", "name", "host_acceptance_rate", "summary", "space", "description", "neighborhood_overview", "notes", "transit",  "access", "interaction", "house_rules", "thumbnail_url", "medium_url", "picture_url",  "xl_picture_url",  "host_id", "host_url", "host_name", "host_since", "host_location", "host_about", "host_response_time", "host_response_rate","host_thumbnail_url", "host_picture_url",  "host_neighbourhood", "host_listings_count", "host_verifications", "host_has_profile_pic", "host_identity_verified", "street","neighbourhood_cleansed", "neighbourhood_group_cleansed", "city", "state", "market", "smart_location", "country_code", "country", "amenities", "calendar_updated", "calendar_last_scraped", "first_review", "last_review", "license", "jurisdiction_names", "require_guest_profile_picture", "requires_license", "is_location_exact", "require_guest_phone_verification", "calculated_host_listings_count"]
review_columns = ["review_scores_value", "review_scores_rating", "review_scores_accuracy", "review_scores_cleanliness", "reviews_per_month", "review_scores_checkin", "review_scores_communication", "review_scores_location", "review_scores_value"]
```

```python
if drop_columns:
    df.drop(columns_to_be_deleted, axis=1, inplace=True)
    df.drop(review_columns, axis=1, inplace=True)
    #Feature Engineering, remove all the occurances of '$' symbol
    drop_columns = False
    
```

```python
if replace_dollar_symbols:
    df.price = df.price.str.replace('$', '').str.replace(',', '').astype(float)
    df.weekly_price = df.weekly_price.str.replace('$', '').str.replace(',', '').astype(float)
    df.monthly_price = df.monthly_price.str.replace('$', '').str.replace(',', '').astype(float)
    df.cleaning_fee = df.cleaning_fee.str.replace('$', '').str.replace(',', '').astype(float)
    df.security_deposit = df.security_deposit.str.replace('$', '').str.replace(',', '').astype(float)
    df.extra_people = df.extra_people.str.replace('$', '').str.replace(',', '').astype(float)
    replace_dollar_symbols = False


```

```python
#Print the percentage of all missing values
(len(df)-df.count())/len(df)*100
```

```python
#We see we really have lot of missing values for the weekly_price and monthly_price column. Similarly the square_feet
df.drop(["weekly_price", "monthly_price", "square_feet"], axis=1, inplace=True)
#has_availability makes no sense, as its just one value in all columns, similar is the case with is_business_travel_ready
df.drop(["has_availability", "is_business_travel_ready"], axis=1, inplace=True)
```

```python
#Generate a Categorical and Numerical Column List
numerical_columns = []
categoical_columns = []
for column in df.columns:
    if df[column].dtype == 'object':
        categoical_columns.append(column)
    else:
        numerical_columns.append(column)
```
